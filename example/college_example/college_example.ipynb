```python
import sys
sys.path.append('../..')
from datamart import search, join
from datamart.utilities.utils import SEARCH_URL
import pandas as pd
```

```python
supplied_df = pd.read_csv('college.csv')
print(supplied_df.head())
```

```python
query = {
    'dataset': {
        'about': 'college, education'
    }, 
    'required_variables': [
        {
            'type': 'dataframe_columns',
            'names': ['INSTNM']
        }
    ]
}
```

```python
res = search(url = SEARCH_URL, 
             query = query, 
             data = supplied_df)
```

```python
cnt = 0
for r in res:
    print('RESULT %d: ' % cnt)
    cnt += 1
    print(r.summary)
```

```python
augmented = join(left_data = supplied_df, 
                 right_data = res[2],
                 left_columns = [[2]],
                 right_columns = [[3]]
                )
```

```python
print(augmented.head(10))
```

```python
with open('augmented.csv', 'w') as fout:
    fout.write(augmented.to_csv(index=False))
```
